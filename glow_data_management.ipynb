```python
#Import and merge all SWOT passes for a river shapefiles
river_rho_shp = merge_shapefiles()
```

```python
export_dataframe(river_rho_shp, is_geodataframe=True)
```

#### 2. Import the GLOW-S files (.parquet and shapefiles)
Using the river ID, both data sets are merged for each river on which the analysis is performed

```python
# Import .gpkg file (This is GLOW-S gpkg file of each river)
river_gpkg = call_file()
river_gpkg
```

```python
# Function to calculate the width median, max and min values, and 2-sigma for each river. Arg: the gpkg (manual data) imported as "river_gpkg"
rio_GLS = process_geopackage(river_gpkg)
rio_GLS
```

```python
# Import the SWOT shapefile for each river (All merged, which means shapefile before any filter or after filtered: 16, 8, 4 bits)
rio_shp_SWOT = call_file()
rio_shp_SWOT
```

```python
# Function to merge the GLOWS gdp and SWOT gpd for each river
"""
The function looks for SWOT observations close to the GLOW-S observation in a threshold of 250m 
(by default but must be modify according with the river width), and filter out the SWOT observations out the threshold
"""
GLWOT = merge_nearest_neighbors(rio_GLS, rio_shp_SWOT, threshold=250)
GLWOT
```

```python
export_dataframe(GLWOT, is_geodataframe=True)
```
